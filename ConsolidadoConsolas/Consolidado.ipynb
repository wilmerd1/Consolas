```python
dataconsolidado ['Registration Year']= pd.DatetimeIndex(dataconsolidado['Registration date']).year
dataconsolidado ['Registration Month']= pd.DatetimeIndex(dataconsolidado['Registration date']).month
dataconsolidado ['Registration Day']= pd.DatetimeIndex(dataconsolidado['Registration date']).day
```

```python
dataconsolidado ['Estimated Solution Year']= pd.DatetimeIndex(dataconsolidado['Estimated Date of Solution']).year
dataconsolidado ['Estimated Solution Month']= pd.DatetimeIndex(dataconsolidado['Estimated Date of Solution']).month
dataconsolidado ['Estimated Solution Day']= pd.DatetimeIndex(dataconsolidado['Estimated Date of Solution']).day
```

```python
dataconsolidado.head(5)
```

```python
datosr = dataconsolidado.loc[(dataconsolidado['Tipo de Caso']=='Requerimientos')]
datosi = dataconsolidado.loc[(dataconsolidado['Tipo de Caso']=='Indicentes')]
```

```python
dfr = datosr.groupby([dataconsolidado['Registration date'].dt.year]).agg('count')
dfi = datosi.groupby([dataconsolidado['Registration date'].dt.year]).agg('count')
```

```python
fig, ax= plt.subplots(1,2, figsize=(15,6))
sns.lineplot(data=dfr, x=dfr['Registration Year'].index, y=dfr['Tipo de Caso'], ax=ax[0])
sns.lineplot(data=dfi, x=dfi['Registration Year'].index, y=dfi['Tipo de Caso'], ax=ax[1])
```
